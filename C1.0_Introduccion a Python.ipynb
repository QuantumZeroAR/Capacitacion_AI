Conversores de valores

```python
print("Bienvenido al conversor de millas a kilómetros")

print("Escribe un número en millas:")
millas = input() #string

print("Tipo de dato de millas", type(millas))
#Convertir de string a numero
millas = float(millas)
print("Tipo de dato de millas", type(millas))

# 1 milla = 1.609 kms
kilometros = millas * 1.609

print("Millas ingresadas:", millas)
print("Kilómetros:", kilometros)
```

Listas

```python
nombres = ["Rodrigo", "Juan", "Pedro", "Santiago", "Jorge", "Raymundo"]
print(nombres)
#f-strings
for i, nombre in enumerate(nombres):
    #print("Se inscribió ", i, "en la lista:", i, nombre)
    print( f"Se inscribió {nombre} en la lista con el índice {i}" )

print( "Bienvenidos a la fiesta", nombres[:3] )
print( "Lo sentimos", nombres[3:])
```
